```python
# Use this cell to begin your analyses, and add as many cells as you would like!
# importing libraries 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#importing datatsets 

lego_sets= pd.read_csv("datasets/lego_sets.csv")
lego_sets.info()

parent_themes = pd.read_csv("datasets/parent_themes.csv")
parent_themes.info()
```

```python
# the # of items in the year column is more than the # of set_numbers - unique ID - taking a closer look

lego_sets

# here we see several NaN values within the dataset
```

```python
#remove NaN values 

lego_sets_clean = lego_sets[lego_sets["set_num"].isna() == False] 

#lego_sets_clean2 = lego_sets_clean.isin(parent_theme)

lego_sets_clean.info()
lego_sets_clean.head()
# now we can begin looking at the lego dataset 
```

```python
# created licensed only parent_themes df & used the .isin() method to filter the lego dataset
lic_parent_themes = parent_themes[parent_themes["is_licensed"]==True]
lic_parent_themes

lego_sets_clean2 = lego_sets_clean[lego_sets_clean["parent_theme"].isin(lic_parent_themes["name"])]
lego_sets_clean2
```

```python
theme_group = lego_sets_clean2["parent_theme"].value_counts(normalize = True).mul(100).sort_values(ascending = False)

theme_group[0:10] 
```

```python
# here we can see that stars wars accounts for roughly 51% of released sets
# we will save this value as "the_force"

the_force = theme_group["Star Wars"].astype('int')
the_force
```

```python

# in what year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)?

pop_parent_themes = lego_sets_clean2[["set_num","parent_theme","year"]]

pivot_themes = pop_parent_themes.pivot_table(index='year', columns='parent_theme', values='set_num', aggfunc='count')
pivot_themes = pivot_themes.fillna(0)                                          
pivot_themes

```
